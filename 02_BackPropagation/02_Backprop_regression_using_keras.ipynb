```python
import numpy as  np 
import pandas as pd 
```

```python
df=pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]] , columns=['cgpa','profile_score','lpa'])
df
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense 
```

```python
model=Sequential()  # empty neural network is created 

model.add(Dense(2,activation='relu',input_dim=2))  # 2:no of neurons  hidden layer 1  , input_dimm =2 : input layer has 2 featues 
model.add(Dense(1,activation='linear')) #1  : neuron in output layer  
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
optimizer=keras.optimizers.Adam(learning_rate=0.001)  # updating weights 
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(df.iloc[:,0:-1].values,df['lpa'].values,epochs=75,verbose=1,batch_size=1)
```

```python
model.get_weights()
```

```python

a = np.array([[5, 12],[8,8]])   # shape = (1, 2)
prediction = model.predict(a)
print(prediction)  # prediction 
```
