```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa', 'profile_score', 'lpa']) # creating our own dataframe
df
```

```python
def initialize_parameters(layer_dims):
  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)         

  for l in range(1, L):

    parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))
      
  return parameters

# we have this function ....f  this will initialize random values to all weights and biases 
# in this we give [2 2 1 ] means 2 intput (cg , profile_score )
# 2 nodes in hidden layer 1 
# 1 node as ouput layer  
# therefore we need total 6 weights and 3 biases  therefore total 9 weights/biases 
# as input we will give [2 2 1]

```

```python
initialize_parameters([2,2,1])
# w1 contains four values i.e weights of 4 lines connecting b/w layer 0 to layer 1  i.e 0.1 
# b1 is biases of 2 nodes at hidden layer 1 or layer 1  i.e 0  
# similarly w2 contains weights of line connecting layer 1 with output node i.e 0.1 
# b2 is bias of output node / perceptron i.e 0 
```

```python
def linear_forward(A_prev, W, b):  # A_prev : previous layer se jo mil rha as input, W and bias me dot product : ....give output from current c
  Z = np.dot(W.T, A_prev) + b
  return Z
```

```python
# Forward Prop
def L_layer_forward(X, parameters):  # particular student/row ke liye prediction krega 
    # Parameters are w and b 
    # X : values of row 

  A = X
  L = len(parameters) // 2                  # number of layers in the neural network
  
  for l in range(1, L+1):
    A_prev = A 
    Wl = parameters['W' + str(l)]
    bl = parameters['b' + str(l)]
    print("A"+str(l-1)+": ", A_prev)
    print("W"+str(l)+": ", Wl)
    print("b"+str(l)+": ", bl)
    print("--"*20)

    A = linear_forward(A_prev, Wl, bl)
    print("A"+str(l)+": ", A)
    print("**"*20)
          
  return A,A_prev

# A_prev : previous layers ka output 
# A0 : input values 
# A1 : output values of layer 1
# A2 : output of last node i.e : predicted value 
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[0][0]

# Parameter initialization
parameters = initialize_parameters([2,2,1])
y_hat,A1 = L_layer_forward(X, parameters)
```

```python
def update_parameters(parameters,y,y_hat,A1,X): # updating parameters 
  # total nine updates ..
  # paramaters : 
  # y : real y value
  # y_hat : predicted value 
  # A1: output values of layer 1 or previous layer
  # X : input values  or data 
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
update_parameters(parameters,y,y_hat,A1,X) # updating parameters 
```

```python
##epochs implementation

parameters = initialize_parameters([2,2,1])
epochs = 5

for i in range(epochs):

  Loss = [] # maintaining loss values 

  for j in range(df.shape[0]):  # visit all rows 

    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1) # Shape(no of features, no. of training example)
    y = df[['lpa']].values[j][0]

    # Parameter initialization


    y_hat,A1 = L_layer_forward(X,parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters,y,y_hat,A1,X)

    Loss.append((y-y_hat)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

parameters  # final values of weights and biases
```
