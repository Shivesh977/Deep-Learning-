```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
from keras.applications.vgg16 import VGG16
```

Feature extraction transfer learning is used 
In this ..whole cnn part is freezed only ann part is removed and replaced by manual ann layers


```python
conv_base = VGG16( #using vgg16 model which has been trained on imagenet 
    weights='imagenet', # we want to use weights .which were found while being trained on imagenet database 
    include_top = False, # only include cnn part ....ann part is made manually 
    input_shape=(150,150,3)
)
```

```python
conv_base.summary() # inlcudes only cnn layers 
```

```python
# Adding self made ann layer
model = Sequential()

model.add(conv_base) # cnn layer 
model.add(Flatten()) # flattening 
model.add(Dense(256,activation='relu')) # ann layer 
model.add(Dense(1,activation='sigmoid')) # output layer 
```
