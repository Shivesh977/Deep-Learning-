```python
from keras.layers import * 
```

```python
# Tell the shape of input layer 
x=Input(shape=(3,)) # as we have  3 inputs salary,height,and martial status
```

```python
# Making 2 Dense layer
hidden1=Dense(128,activation='relu')(x) # first dense layer will get input from x ......128 neurons
hidden2=Dense(64,activation='relu')(hidden1) # second dense layer will get input from hidden1

# uptill now we have linear pipeline now branching 
# Branching 

output1= Dense(1,activation='linear')(hidden2) # regression problem age 
output2=Dense(1,activation='sigmoid')(hidden2) # classification problm city:Delhi/Mumbai

```

Non linear nn with 2 input and 1 output ...shown in image 01.2.png

```python
from keras.layers import *
from keras.models import Model
```

```python
# define two sets of inputs
inputA = Input(shape=(32,))
inputB = Input(shape=(128,))
```

```python
# the first branch operates on the first input
x = Dense(8, activation="relu")(inputA)
x1 = Dense(4, activation="relu")(x)
```

```python
# the second branch opreates on the second input
y = Dense(64, activation="relu")(inputB)
y1 = Dense(32, activation="relu")(y)
y2 = Dense(4, activation="relu")(y1)
```

```python
# combine the output of the two branches
combined = concatenate([x1, y2])
```

```python
# apply a FC layer and then a regression prediction on the
# combined outputs
z = Dense(2, activation="relu")(combined)
z1 = Dense(1, activation="linear")(z)
```

```python
# our model will accept the inputs of the two branches and
# then output a single value
model = Model(inputs=[inputA, inputB], outputs=z1)
```

```python
from keras.utils import plot_model
plot_model(model)
```
