```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Create tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')  
# oov_token handles unseen words


tokenizer.fit_on_texts(docs)

tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences

sequences = pad_sequences(sequences,padding='post') # padding is done to make every sequence of same size ....
```

Using imbd dataset for sentiment analysis

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train[0]
```

```python
len(X_train[2])
```

```python
X_train = pad_sequences(X_train,padding='post')
X_test = pad_sequences(X_test,padding='post')
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False)) #32 neurons....
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

Using embedding 

```python
from keras.datasets import imdb
# from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post')
X_test = pad_sequences(X_test,padding='post')
```

```python
model = Sequential()
model.add(Embedding(10000, 2,50)) # 10000 unique words in corpus , 
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```
