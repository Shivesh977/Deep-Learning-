```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data() # loading dataset 
```

```python
from tensorflow.keras.layers import Conv2D, AveragePooling2D
```

```python
# Making LeNet CNN architecture
model = Sequential()

# First layer
model.add(Conv2D(6,kernel_size=(5,5),padding='valid',activation='tanh',input_shape=(32,32,1))) # 6 filters each of size 5*5 
model.add(AveragePooling2D(pool_size=(2, 2), strides=2, padding='valid'))#max pooling 

# Second layer
model.add(Conv2D(16,kernel_size=(5,5),padding='valid',activation='tanh')) # 6 filters each of size 5*5 
model.add(AveragePooling2D(pool_size=(2, 2), strides=2, padding='valid'))#max pooling 

model.add(Flatten()) # flattening the model 

# Adding the ANN architecture
model.add(Dense(120,activation='tanh'))
model.add(Dense(84,activation='tanh'))
model.add(Dense(10,activation='softmax'))
```
